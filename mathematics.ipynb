# Pi value

```python
Here,I have used montecarlo method to find the pi value.
```

```python
from simple_colors import *
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import random
circlex=[]
circley=[]
squarex=[]
squarey=[]
n=10000 # number of darts that we are throwing
i=1
while i<n:
    x=random.uniform(-1,1) # we are assuming circle of radius 1,you can choose any number
    y=random.uniform(-1,1)
    if x**2 + y**2<=1: # circle condition
        circlex.append(x)
        circley.append(y)
    else:
        squarex.append(x)
        squarey.append(y)
    i=i+1
pi=4*len(circlex)/n # here circlex=circley
print(red(f'the pi value of our experiment = {pi}','bold'))
print(green(f'the actual value of pi = {np.pi}','bold'))
plt.plot(circlex,circley,'b.')
plt.plot(squarex,squarey,'g.');
```

# Wallis integral

```python
import numpy as np
def wallis_pi(n):
    """Walis integral for calculating pi value for infinite series"""
    c=1
    i=1
    while(i<n):
        temp=(4*i**2)/(4*i**2-1)
        c=temp*c
        i+=1
    return 2*c


print(f' actual pi value {np.pi}')
print(f' pi value calculated from wallis integral {wallis_pi(1000000) }')
       
```

# Ramanujan infinite series integral

```python
import numpy as np
def fact(n):
    if n<0:
        print('please enter positive number')
    elif n==1 or n==0:
        return 1
    else:
        return n*fact(n-1)
    
def rama_integral(n):
    k=0
    val=0
    c=2*np.sqrt(2)/9801
    while(k<n):
        temp=c*((fact(4*k))*(1103+26390*k))/(((fact(k))**4)*396**(4*k))
        val+=temp
        k+=1
    return 1/val
        
rama_integral(22)        
```

```python
import random
from simple_colors import*
def roll_dice():
    roll=random.randint(1,100)
    if  1<roll<=50:
        #print(red(f'{roll}, roll is between 1-50,you lose,play again!','bold'))
        return False
    else:
        #print(blue(f'{roll}, roll is between 51-100.you win','bold'))
        return True
        
        
def simple_battor(funds,initial_wages,counter):
    value=funds
    c=0
    c1=0
    c2=0
    c3=0
    while(c<counter):
        if roll_dice():
            value+=initial_wages
            
        
        else:
            value-=initial_wages
            
        #print(f' funds {value}')
        if funds<value:
            c1+=1
            print(cyan(f' win {value}','bold'))
        elif funds==value:
            c3+=1
            print(f'draws {value}')
        elif value<=0:
            
            print('broke')
        else:
            c2+=1
            print(red(f' lose {value}','bold'))
        c+=1
    print('*'*30)
    print('gambling summary'.upper())
    print('*'*30)
    print(f' you win {c1} and loose {c2}  draws {c3}')
        
    
simple_battor(10000,100,100)
```
