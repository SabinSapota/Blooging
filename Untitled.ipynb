# Global Question

If I plot all these country in same figure then it will looks messy hence I take most prominent country from each continent
and compare with orther country for the given index parameter as shown below

```python
def plot(C,parameter1,parameter2,C1,C2,C3,C4,C5):
    df1=df[df['Country Name']==C]
    df1.drop(['Country ISO3','Indicator Id'],axis=1,inplace=True)
    df1=df1.set_index(['Country Name','Subindicator Type','Indicator'])
    
    df2=df[df['Country Name']==C1]
    df2.drop(['Country ISO3','Indicator Id'],axis=1,inplace=True)
    df2=df2.set_index(['Country Name','Subindicator Type','Indicator'])
    
    df3=df[df['Country Name']==C2]
    df3.drop(['Country ISO3','Indicator Id'],axis=1,inplace=True)
    df3=df3.set_index(['Country Name','Subindicator Type','Indicator'])
    
    df4=df[df['Country Name']==C3]
    df4.drop(['Country ISO3','Indicator Id'],axis=1,inplace=True)
    df4=df4.set_index(['Country Name','Subindicator Type','Indicator'])
    
    df5=df[df['Country Name']==C4]
    df5.drop(['Country ISO3','Indicator Id'],axis=1,inplace=True)
    df5=df5.set_index(['Country Name','Subindicator Type','Indicator'])
    
    df6=df[df['Country Name']==C5]
    df6.drop(['Country ISO3','Indicator Id'],axis=1,inplace=True)
    df6=df6.set_index(['Country Name','Subindicator Type','Indicator'])
    
    
    
    import seaborn as sns
    sns.set_style("darkgrid",{'axes.grid' : False})
    plt.figure(figsize=(15,7))
    plt.scatter(df1.loc[C,parameter1,parameter2].index,df1.loc[C,parameter1,parameter2].values,marker='o',color='black',label=C)
    plt.scatter(df2.loc[C1,parameter1,parameter2].index,df2.loc[C1,parameter1,parameter2].values,marker='o',color='red',label=C1)
    plt.scatter(df3.loc[C2,parameter1,parameter2].index,df3.loc[C2,parameter1,parameter2].values,marker='o',color='green',label=C2)
    plt.scatter(df4.loc[C3,parameter1,parameter2].index,df4.loc[C3,parameter1,parameter2].values,marker='o',color='#9D2933',label=C3)
    plt.scatter(df5.loc[C4,parameter1,parameter2].index,df5.loc[C4,parameter1,parameter2].values,marker='o',color='#F8DE7E',label=C4)
    plt.scatter(df6.loc[C5,parameter1,parameter2].index,df6.loc[C5,parameter1,parameter2].values,marker='o',color='#FFB6C1',label=C5)
    
    
    
    plt.title(f'{parameter2} for {C},{C1},{C2},{C3},{C4} and {C5}',fontweight='bold',fontsize=20,color='blue')
    plt.xlabel('year',fontweight='bold',fontsize=15)
    plt.ylabel(f'{parameter2}',fontweight='bold',fontsize=15);
    plt.legend()
    
    #return df1.loc[C,parameter1,parameter2]
plot('Nepal','Index','Overall Global Gender Gap Index','China','France','India','Australia','Canada')
    
```

From above figure we can conclude that France and canda has better overall Global gender gap index and Nepal and India has least
index.But, we also see that both Nepal and India global gender gap index goes  more or less increasing with increase in year.
Both country are imporving each year

```python
plot('Nepal','Index','Global Gender Gap Political Empowerment subindex','China','France','India','Australia','Canada')
```

Here,Global gender gap political empowerment subindex for  Nepal in inreasing with increasing in year.where as for china it decreasing.
France is at better position in few past year for this subindex.

```python
plot('Nepal','Index','Global Gender Gap Health and Survival Subindex','China','France','India','Australia','Canada')
```

France and Canada are at doing good for global gender gap health and Survival subindex.where as India and China are at bottom of the

# Comparing Subsets

Now,I will compare Nepal with its neighbouring country India on the same parameter

```python
def plot(C,parameter1,parameter2,C1):
    df1=df[df['Country Name']==C]
    df1.drop(['Country ISO3','Indicator Id'],axis=1,inplace=True)
    df1=df1.set_index(['Country Name','Subindicator Type','Indicator'])
    
    df2=df[df['Country Name']==C1]
    df2.drop(['Country ISO3','Indicator Id'],axis=1,inplace=True)
    df2=df2.set_index(['Country Name','Subindicator Type','Indicator'])
    
    
    import seaborn as sns
    sns.set_style("darkgrid",{'axes.grid' : False})
    plt.figure(figsize=(15,7))
    plt.scatter(df1.loc[C,parameter1,parameter2].index,df1.loc[C,parameter1,parameter2].values,marker='o',color='black',label='Nepal')
    plt.scatter(df2.loc[C1,parameter1,parameter2].index,df2.loc[C1,parameter1,parameter2].values,marker='o',color='red',label='India')
    plt.title(f'{parameter2} for {C} and {C1}',fontweight='bold',fontsize=20,color='blue')
    plt.xlabel('year',fontweight='bold',fontsize=15)
    plt.ylabel(f'{parameter2}',fontweight='bold',fontsize=15);
    plt.legend()
    
    #return df1.loc[C,parameter1,parameter2]
plot('Nepal','Index','Overall Global Gender Gap Index','India')
    
```
