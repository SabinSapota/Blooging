```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style("darkgrid",{'axes.grid' : False})
```

```python
df=pd.read_csv('anscombe.tsv',delimiter='\t')
df
```

```python
first_dataset=df[df.dataset==1]
first_dataset
```

```python
def drawplot(x):
    dataset=df[df.dataset==x]
    
    plt.figure(figsize=(10,6))
    plt.scatter(dataset.x,dataset.y,color='red',s=100,edgecolors='black')
    plt.xlabel('x cordinate',fontsize=20)
    plt.ylabel('y cordinate',fontsize=20)
    plt.title(f'scatter plot for {x} data set',fontsize=25);
```

```python
drawplot(1)
```

```python
drawplot(2)
```

```python
drawplot(3)
```

```python
drawplot(4)
```
