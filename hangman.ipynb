# hangmaan game

This is hangmaan game with a little twist.
This game gives you hint to guess the word

```python
from simple_colors import*
import random
print(red(f' welcome to hangmaan game'.upper(),'bold'))
print('*'*25)
name=input('what is your name: ')
print(green(f' hello ,{name.title()}  .....','bold'))
print('Starting guessing.......')
d = {'VENEZUELA':'CARACAS', 'CANADA':'TORONTO','NEPAL':'Kathmandu','INDia':'Dheli','france':'paris','chiana':'beijing'}

def fxn():
    country, capital = random.choice(list(d.items()))
    print(red(f' hint:- it is capital of {country}','bold'))
    turns=3
    guesses=''
    while turns>0:
        failed=0
        for char in capital.lower():
            if char in guesses:
                print(char)
            else:
                print('_')
                failed+=1
        if failed==0:
            print(green(f'Congratulation {name}!, you have won the game by guessing correctly','bold'))
            break
        guess=input('guess the character: ')
        guesses+=guess
        if guess not in capital.lower():
            turns-=1
            print(red('wrong guess','bold'))
            print(cyan(f' you have {turns} guess now ','bold'))
            print('*'*25)
        if turns==0:
            print(red('game is over , you loss the game','bold'))
            
    check=input('Do You want to play y/n ?')
    if check=='y':
        fxn()
   
        


fxn()
            
    
```
